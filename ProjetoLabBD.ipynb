```
# Configurando o ambiente para a análise dos dados
import math
import pandas as pd
import seaborn as sb
import matplotlib.pyplot as plt
low_memory = False
%matplotlib inline
pd.options.display.max_columns = 80
pd.options.display.max_rows = 90
```

```
def montarPizza(dicicionarioGraficoMontar, nomeArquivo): # Gráfico pizza
    plt.axis('equal')
    plt.pie(
        dicicionarioGraficoMontar['y'],
        labels = dicicionarioGraficoMontar['x'],
        autopct = '%1.1f%%'
    )
    plt.title(dicicionarioGraficoMontar['titulo'])
    plt.autoscale()
    plt.tight_layout()
    # Descomente este bloco se você estiver usando o Jurpyter Notebook Local
    # e comente o bloco que usa a Drive do Google
    # plt.savefig('Gráficos/' + nomeArquivo + '.png', bbox_inches='tight')
    plt.savefig('/content/drive/My Drive/Colab Notebooks/Gráficos/' + nomeArquivo + '.png', bbox_inches='tight')
    plt.show()

def montarGrafico(dicicionarioGraficoMontar, nomeArquivo): # Montar gráfico linhas
    fig, ax = plt.subplots()
    ax.plot(dicicionarioGraficoMontar['x'], dicicionarioGraficoMontar['y'], color = dicicionarioGraficoMontar['corLinha'])
    plt.scatter(dicicionarioGraficoMontar['x'], dicicionarioGraficoMontar['y'], color = dicicionarioGraficoMontar['corPontos'])
    plt.title(dicicionarioGraficoMontar['titulo'])
    plt.xlabel(dicicionarioGraficoMontar['xLabel'])
    plt.ylabel(dicicionarioGraficoMontar['yLabel'])
    plt.autoscale()
    plt.tight_layout()
    # Descomente este bloco se você estiver usando o Jurpyter Notebook Local
    # e comente o bloco que usa a Drive do Google
    # plt.savefig('Gráficos/' + nomeArquivo + '.png', bbox_inches='tight')
    plt.savefig('/content/drive/My Drive/Colab Notebooks/Gráficos/' + nomeArquivo + '.png', bbox_inches='tight')
    plt.show()

def montarGraficoBarras(dicicionarioGraficoMontar, nomeArquivo): # Montar gráfico de barras
    fig, ax = plt.subplots()
    ax.bar(dicicionarioGraficoMontar['x'], dicicionarioGraficoMontar['y'])
    plt.title(dicicionarioGraficoMontar['titulo'])
    plt.xlabel(dicicionarioGraficoMontar['xLabel'])
    plt.ylabel(dicicionarioGraficoMontar['yLabel'])
    plt.autoscale()
    plt.tight_layout()
    # Descomente este bloco se você estiver usando o Jurpyter Notebook Local
    # e comente o bloco que usa a Drive do Google
    # plt.savefig('Gráficos/' + nomeArquivo + '.png', bbox_inches='tight')
    plt.savefig('/content/drive/My Drive/Colab Notebooks/Gráficos/' + nomeArquivo + '.png', bbox_inches='tight')
    plt.show()

# intervalo = (x1, y1)
# xytxt = (x2, y2)
def montarGraficoComentar(dicicionarioGraficoMontar, intervalo, xytxt, comentario, nomeArquivo): # Gráfico de barras com comentário
    fig, ax = plt.subplots()
    ax.bar(dicicionarioGraficoMontar['x'], dicicionarioGraficoMontar['y'])
    ax.annotate(
        comentario,
        xy = intervalo,
        xycoords = 'data',
        xytext = xytxt,
        textcoords = 'data',
        arrowprops = dict(arrowstyle = "->", connectionstyle = "arc3")
    )
    plt.title(dicicionarioGraficoMontar['titulo'])
    plt.xlabel(dicicionarioGraficoMontar['xLabel'])
    plt.ylabel(dicicionarioGraficoMontar['yLabel'])
    # Descomente este bloco se você estiver usando o Jurpyter Notebook Local
    # e comente o bloco que usa a Drive do Google
    # plt.savefig('Gráficos/' + nomeArquivo + '.png', bbox_inches='tight')
    # plt.savefig('/content/drive/My Drive/Colab Notebooks/Gráficos/' + nomeArquivo + '.png')
    plt.show()
```

```
count_ext, count_mob, count_pes, count_est, count_bol, count_mon = 0, 0, 0, 0, 0, 0
for i in range(len(data[2016]["content"])):
  if data[2016]["content"]['IN_COMPL_EXTENSAO'][i] == 1: count_ext += 1
  if data[2016]["content"]['IN_MOBILIDADE_ACADEMICA'][i] == 1: count_mob += 1
  if data[2016]["content"]['IN_BOLSA_PESQUISA'][i] == 1: count_pes += 1
  if data[2016]["content"]['IN_BOLSA_ESTAGIO'][i] == 1: count_est += 1
  if data[2016]["content"]['IN_BOLSA_EXTENSAO'][i] == 1: count_bol += 1
  if data[2016]["content"]['IN_BOLSA_MONITORIA'][i] == 1: count_mon += 1

data[2016]["values"]["IN_COMPL_EXTENSAO_total"] = count_ext
data[2016]["values"]["IN_COMPL_EXTENSAO_per_cent"] = (count_ext / (len(data[2016]["content"])))

data[2016]["values"]["IN_MOBILIDADE_ACADEMICA_total"] = count_mob
data[2016]["values"]["IN_MOBILIDADE_ACADEMICA_per_cent"] = (count_mob / (len(data[2016]["content"])))

data[2016]["values"]["IN_BOLSA_PESQUISA_total"] = count_pes
data[2016]["values"]["IN_BOLSA_PESQUISA_per_cent"] = (count_pes / (len(data[2016]["content"])))

data[2016]["values"]["IN_BOLSA_ESTAGIO_total"] = count_est
data[2016]["values"]["IN_BOLSA_ESTAGIO_per_cent"] = (count_est / (len(data[2016]["content"])))

data[2016]["values"]["IN_BOLSA_EXTENSAO_total"] = count_bol
data[2016]["values"]["IN_BOLSA_EXTENSAO_per_cent"] = (count_bol / (len(data[2016]["content"])))

data[2016]["values"]["IN_BOLSA_MONITORIA_total"] = count_mon
data[2016]["values"]["IN_BOLSA_MONITORIA_per_cent"] = (count_mon / (len(data[2016]["content"])))

data[2016]["description"] = (
    ("%5.2f" %(data[2016]["values"]["IN_COMPL_EXTENSAO_per_cent"] * 100 )) + 
    "% dos alunos realizaram algum tipo de atividade extracurricular em 2016.\n" +
    ("%5.2f" %(data[2016]["values"]["IN_MOBILIDADE_ACADEMICA_per_cent"] * 100 )) + 
    "% dos alunos participaram de algum programa de mobilidade acadêmica em 2016.\n" +
    ("%5.2f" %(data[2016]["values"]["IN_BOLSA_PESQUISA_per_cent"] * 100 )) + 
    "% dos alunos receberam algum tipo de bolsa por pesquisas em 2016.\n" +
    ("%5.2f" %(data[2016]["values"]["IN_BOLSA_ESTAGIO_per_cent"] * 100 )) + 
    "% dos alunos receberam algum tipo de bolsa por atividade de estágio em 2016.\n" +
    ("%5.2f" %(data[2016]["values"]["IN_BOLSA_EXTENSAO_per_cent"] * 100 )) + 
    "% dos alunos receberam algum tipo de bolsa por atividades extracurriculares em 2016.\n" +
    ("%5.2f" %(data[2016]["values"]["IN_BOLSA_MONITORIA_per_cent"] * 100 )) + 
    "% dos alunos receberam algum tipo de bolsa por atividades de monitoria em 2016.\n"
)
print(data[2016]["description"])
```
